# Checking Variance Inflation Factor for checking MultiCollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Features']=x.columns
    vif['VIF Value']=[variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
    return vif
```

```python
for i in range(0,100):
    x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2,random_state=i)
    lr.fit(x1_train,y1_train)
    predlr=lr.predict(x1_test)
    lracc=accuracy_score(y1_test,predlr)
    print('At Random state ',i,' Accuracy score is ',lracc*100)
```

```python
maxacc=0
maxrs=0
for i in range(1,100):
    x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.30,random_state=i)
    lr.fit(x1_train,y1_train)
    predlr=lr.predict(x1_test)
    acclr=accuracy_score(y1_test,predlr)
    if acclr>maxacc:
        maxacc=acclr
        maxrs=i
print('The Maximum Accuracy is ',maxacc*100,' at random state ',maxrs)
```
