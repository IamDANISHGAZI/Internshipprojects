```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Features']=x.columns
    vif['VIF Value']=[variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
    return vif
```

# Model Building

```python
# From now on we will be working on our model building, here we are first taking linear Regression since we are dealing with cotinious data and checking r2 score as well.  
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
lm=LinearRegression()
```

```python
# Selecting best Random State for training and testng
for i in range(0,100):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=i)
    lm.fit(xtrain,ytrain)
    predlm=lm.predict(xtest)
    lmacc=r2_score(ytest,predlm)
    print('At Random state ',i,' Accuracy is ',lmacc*100)
    print("\n")
```

```python
# Since 37 is our best random state, for final modelling at random state 60
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=60)
```

```python
# Selecting the best CV value (Cross validation helps us to find underfitting and overfitting of model)
from sklearn.model_selection import cross_val_score
for i in range(2,10):
    cv_score=cross_val_score(lm,x,y,cv=i)*100
    cv_mean=cv_score.mean()
    print(f'At cv value {i} the cross val score is {cv_mean} and  accuracy is ',r2_score(ytest,predlm)*100)
    print('\n')
```
