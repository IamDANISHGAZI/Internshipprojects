```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
svc=SVC()
```

```python
# CHecking Best random state
```

```python
for i in range(0,100):
    x1train,x1test,y1train,y1test=train_test_split(x1,y1,test_size=0.2,random_state=i)
    svc.fit(x1train,y1train)
    predsvc=svc.predict(x1test)
    svcacc=accuracy_score(y1test,predsvc)
    print('At random state ',i,' Accuracy is ',svcacc*100)
```

```python
# Finding best parameter using GridSearchCV
```

```python
parameter={'kernel':['linear','poly','rbf']}
gscvsvc=GridSearchCV(svc,parameter)
gscvsvc.fit(x1train,y1train)
print(gscvsvc.best_params_)
```

```python
svc=SVC(kernel='rbf')
```

```python
x1train,x1test,y1train,y1test=train_test_split(x1,y1,test_size=0.2,random_state=34)
svc.fit(x1train,y1train)
predsvc=svc.predict(x1test)
svcacc=accuracy_score(y1test,predsvc)
print('At random state 34 Accuracy is ',svcacc*100)
```

```python
print('Confusion matrix\n',confusion_matrix(y1test,predsvc))
print('Classification report \n',classification_report(y1test,predsvc))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in range(2,10):
    cvsvc=cross_val_score(svc,x1,y1,cv=i)
    cvsvc_mean=cvsvc.mean()
    print('At cv ',i,' cross val score is ',cvsvc_mean*100,' with accuracy ',svcacc*100)
```
