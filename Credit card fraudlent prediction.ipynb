# Checking the Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Variables']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
```

```python
cal_vif(x)
```

```python
# Dataset doesnt suffer from multicollinearity
```

# 1st model Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve,f1_score
```

```python
lr=LogisticRegression()
```

```python
# Finding Best Random State
for i in range(0,100):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(xtrain,ytrain)
    predlr=lr.predict(xtest)
    lracc=accuracy_score(ytest,predlr)
    print('At Random State ',i,' Accuracy is ',lracc*100)
```

```python
# We are getting Best Accuracy at random state 23
```

```python
# Now Lets see the other scores like auc_roc_score and confusion matrix to get an idea of true positive,false positive,true negative,false negative and classification report
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=23)
lr.fit(xtrain,ytrain)
predlr=lr.predict(xtest)
lracc=accuracy_score(ytest,predlr)
print('At Random State 23 Accuracy is ',lracc*100)
print('F1 score is ',f1_score(ytest,predlr,average='binary')*100)
print('ROC AUC Score is',roc_auc_score(ytest,predlr)*100)
print('Confusion matrix is\n',confusion_matrix(ytest,predlr))
print('Classification report is\n',classification_report(ytest,predlr))
```

```python
# Checking The Best CV
```

```python
for i in range(2,10):
    cvlr=cross_val_score(lr,x,y,cv=i)
    cvlr_mean=cvlr.mean()
    print('At CV ',i,'cross val score is',cvlr_mean*100,'with accuracy',lracc*100)
```

```python
# 9 is the best CV value
```

```python
cvlr=cross_val_score(lr,x,y,cv=9)
cvlr_mean=cvlr.mean()
print('At CV 9 cross val score is',cvlr_mean*100,'with accuracy',lracc*100)
```

# 2nd Model Decision Tree classifiers

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
# Finding best Parameters
parameter={'criterion':['gini','entropy','log_loss'],'max_features':['auto','sqrt','log2']}
gscvdtc=GridSearchCV(dtc,parameter)
gscvdtc.fit(xtrain,ytrain)
print(gscvdtc.best_params_)
```

```python
# The best parameters we get using Hyper parameter tuning for decision tree classifier are criterion= log_loss and max features= auto
```

```python
dtc=DecisionTreeClassifier(criterion='log_loss',max_features='auto')
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=23)
dtc.fit(xtrain,ytrain)
preddtc=dtc.predict(xtest)
dtcacc=accuracy_score(ytest,preddtc)
print('At Random state 23 Accuracy is ',dtcacc*100)
print('ROC AUC SCORE is ',roc_auc_score(ytest,preddtc)*100)
print('F1 score is ',f1_score(ytest,preddtc,average='binary')*100)
print('Confusion matrix\n',confusion_matrix(ytest,preddtc))
print('Classification report\n',classification_report(ytest,preddtc))    
```

```python
# Now Checking it with cv
cvdtc=cross_val_score(dtc,x,y,cv=9)
cvdtc_mean=cvdtc.mean()
print('At CV 9 cross val score is ',cvdtc_mean*100,'with Accuracy ',dtcacc*100)
```
