```python
# Now Lets Check for MultiCollinearity
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Variable']=x.columns
    vif['VIF']=[variance_inflation_factor(x,i)for i in range(x.shape[1])]
    return(vif)
```

```python
cal_vif(x)
```

```python
# From The Above fuction it is clear thar RA, ER, and ERA suffer From multi collinearity so removing these Columns
```

```python
x.drop(['ER'],axis=1,inplace=True)
```

```python
cal_vif(x)
```

```python
x.drop(['RA'],axis=1,inplace=True)
```

```python
cal_vif(x)
```

```python
# Now There is no multicollinearity
```

```python
# Now Moving ahead for our First Model Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score
```

```python
lr=LinearRegression()
```

```python
# Finding Best Random State
```

```python
for i in range(0,100):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(xtrain,ytrain)
    predlr=lr.predict(xtest)
    lracc=r2_score(ytest,predlr)
    print('At Random State ',i,' Accuracy is ',lracc*100)
```

```python
# We are getting best Accuracy at Random State 47
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=47)
lr.fit(xtrain,ytrain)
predlr=lr.predict(xtest)
lracc=r2_score(ytest,predlr)
print('At Random State 47 Accuracy is ',lracc*100)
```

```python
# Now Checking it with cross validation
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# Finding best cv value
```

```python
for i in range(2,10):
    cvlr=cross_val_score(lr,x,y,cv=i)
    cvlr_mean=cvlr.mean()
    print('At CV ',i,' Cross Val Score is ',cvlr_mean*100,' wih Linear Regression Accuracy ',lracc*100)
```

```python
# 4 is Best CV
```

```python
cvlr=cross_val_score(lr,x,y,cv=4)
cvlr_mean=cvlr.mean()
print('At CV 4 Cross Val Score is ',cvlr_mean*100,' wih Linear Regression Accuracy ',lracc*100)
```
