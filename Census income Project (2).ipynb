```python
# Now Checking For Multicollinearity
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Variable']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return vif
```

```python
cal_vif(x)
```

```python
# It is clear few columns suffer Multicollinearity so removing them 
```

# Model Building

```python
# Now Our Data Cleaning Process is complete and now apply our First Model Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,roc_auc_score,f1_score
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
```

```python
# Finding Best Random State
```

```python
lr=LogisticRegression()
for i in range(0,100):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(xtrain,ytrain)
    predlr=lr.predict(xtest)
    lracc=accuracy_score(ytest,predlr)
    print('At Random state ',i,' Accuracy is ',lracc*100)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=41)
lr.fit(xtrain,ytrain)
predlr=lr.predict(xtest)
lracc=accuracy_score(ytest,predlr)
print('At Random state 41 Accuracy is ',lracc*100)
```

```python
print('Classification report\n',classification_report(ytest,predlr))
print('Confusion matrix\n',confusion_matrix(ytest,predlr))
print('AUC_ROC Score is',roc_auc_score(ytest,predlr)*100)
print('F1 score is ',f1_score(ytest,predlr)*100)
```

```python
# Now Checking its CV
```

```python
for i in range(2,10):
    cvlr=cross_val_score(lr,x,y,cv=i)
    cvlr_mean=cvlr.mean()
    print('At CV ',i,' cross val score is ',cvlr_mean*100,' with Logistic Regression Accuracy ',lracc*100)
```

```python
# 8 is best CV
```

```python
cvlr=cross_val_score(lr,x,y,cv=8)
cvlr_mean=cvlr.mean()
print('At CV 8 cross val score is ',cvlr_mean*100,' with Logistic Regression Accuracy ',lracc*100)
```

```python
# Now lets see the ROC curve 
```

```python
fpr,tpr,_=roc_curve(ytest,predlr)
plt.plot(fpr,tpr,linestyle='--',marker='.')
plt.title('ROC Curve')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.show()
```
