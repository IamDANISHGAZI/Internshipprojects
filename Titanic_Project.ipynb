```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Checking MultiCollinearity
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['variables']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
# First using classification model Logistic regression and checking confusion matrix and classification report along with accuracy score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,roc_auc_score,f1_score
from sklearn.model_selection import train_test_split
```

```python
lr=LogisticRegression()
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    lracc=accuracy_score(y_test,predlr)
    print('Accuracy is ',lracc*100,' at Random State ',i)
    print('\n')
    
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=75)
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
lracc=accuracy_score(y_test,predlr)
print('Accuracy is ',lracc*100,' at Random State 75')
```

```python
print(confusion_matrix(y_test,predlr))
```

```python
print(classification_report(y_test,predlr))
```

```python
# Checking f1 score,confusion matrix, auc_roc score and classification report and then plotting AUC_ROC curve
```

```python
print('ROC AUC score ',roc_auc_score(y_test,predlr))
print('f1 score ',f1_score(y_test,predlr))
print('Confusion matrix\n',confusion_matrix(y_test,predlr))
print('Classification report\n',classification_report(y_test,predlr))
```

```python
tpr,fpr,_=roc_curve(y_test,predlr)
plt.plot(tpr,fpr,ls='--',marker='.')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
# Checking cross val score
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in range(2,10):
    cvlr=cross_val_score(lr,x,y,cv=i)
    cvlr_mean=cvlr.mean()
    print('At cv ',i,' Cross val scor is ',cvlr_mean*100,' with Logistic Regression Accuracy ',lracc*100)
    print('\n')
```

```python
cvlr=cross_val_score(lr,x,y,cv=8)
cvlr_mean=cvlr.mean()
print('At cv 8 Cross val score is ',cvlr_mean*100,' with Logistic Regression Accuracy ',lracc*100)
```
