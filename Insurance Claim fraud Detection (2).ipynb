```python
# Now lets Quantify Outliers
```

```python
def outliers(dt,fs):
    Q1=dt[fs].quantile(0.25)
    Q3=dt[fs].quantile(0.75)
    IQR=Q3-Q1
    
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    
    ls=dt.index[(dt[fs]<lower_bound)|(dt[fs]>upper_bound)]
    return ls
```

```python
ind_list=[]
for feature in ['age','policy_annual_premium','umbrella_limit','total_claim_amount','property_claim','fraud_reported']:
    ind_list.extend(outliers(df,feature))
```

```python
def remove(dt,ls):
    ls=sorted(set(ls))
    dt=dt.drop(ls)
    return dt
```

```python
df_cleaned=remove(df,ind_list)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Variables']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return vif
```
